```python
def replace(x) :
    return ''.join([i for i in x if i not in punctuation])
```

```python
matches['venue'] = matches['venue'].apply(replace)
matches['venue'] = matches['venue'].apply(lambda x: x.replace('Punjab Cricket Association IS Bindra Stadium Mohali','Punjab Cricket Association Stadium Mohali'))
venues= list(matches.venue.value_counts().index)
venues
```

```python
def balls(cols) :
    x = cols[0]
    y = cols[1]
    if x > 0 or y > 0 :
        return 0
    else :
        return 1
def out_bowler(x) :
    if x in ('caught','bowled','lbw','stumped','caught and bowled','hit wicket') :
        return 1
    else :
        return 0
def team(x) :
    if x in CSK :
        return 'CSK'
    elif x in RCB :
        return 'RCB'
    elif x in KKR :
        return 'KKR'
    elif x in SRH :
        return 'SRH'
    elif x in MI :
        return 'MI'
    elif x in DC :
        return 'DC'
    elif x in RR :
        return 'RR'
    elif x in KXIP :
        return 'KXIP'
    else :
        return 'None'
    
def is_played(cols):
    x = cols[0]
    y = cols[1]
    if x >= 20:
        return 1
    elif  x < 20 and x >= 10 and y==0 :
        return 1
    else :
        return 0
    
def fillna(cols) :
    x1 = cols[0]
    x2 = cols[1]
    x3 = cols[2]
    return (x1+x2+x3)/3

def fillna2(cols) :
    x1 = cols[0]
    x2 = cols[1]
    return (x1+x2)/2
```

```python
total_df = pd.merge(left=deliveries,right=matches,left_on='match_id',right_on='id')
total_df['balls'] = total_df[['wide_runs','noball_runs']].apply(balls,axis=1)
total_df['out_bowler'] = total_df['dismissal_kind'].apply(out_bowler)
total_df.drop(labels=['wide_runs','bye_runs','legbye_runs','noball_runs','penalty_runs','id','player_dismissed','dismissal_kind','extra_runs','total_runs'],axis=1,inplace=True)
total_df['Own_Team'] = total_df['batsman'].apply(team)
total_df['Opp_Team'] = total_df['bowler'].apply(team)
total_df = total_df.groupby(by=['match_id','batsman','bowler','venue','Own_Team','Opp_Team'],as_index=False).sum()[['match_id','batsman','bowler','batsman_runs','venue','balls','out_bowler','Own_Team','Opp_Team']]
total_df = total_df[(total_df['Own_Team']!='None') & (total_df['Opp_Team']!='None') & (total_df['Opp_Team']!= total_df['Own_Team'])]
```

```python
def batting_per(batting_score,player,own_team,opp_team,match_venue,teams,venues) :
    df = batting_score[batting_score['batsman']==player]
    df = df.sort_values(by='match_id')
    df['pre1'] = df['batsman_runs'].shift(1)
    df['pre2'] = df['batsman_runs'].shift(2)
    df['pre3'] = df['batsman_runs'].shift(3)
    df.dropna(inplace=True)
    opp_df = pd.DataFrame()
    for i in [i for i in teams if i not in own] :
        opp = batting_score[(batting_score['batsman']==player) & (batting_score['Opp_Team']==i)]
        opp = opp.sort_values(by='match_id')
        opp['pre_opp1'] = opp['batsman_runs'].shift(1)
        opp['pre_opp2'] = opp['batsman_runs'].shift(2)
        opp['pre_opp3'] = opp['batsman_runs'].shift(3)
        opp_df = opp_df.append(opp)
    opp_df = opp_df[['match_id','batsman','Opp_Team','venue','pre_opp1','pre_opp2','pre_opp3']]
    bat1 = pd.merge(df,opp_df,left_on=['match_id','batsman','Opp_Team','venue'],right_on=['match_id','batsman','Opp_Team','venue'],how='left')
    venue_df = pd.DataFrame()
    for j in venues :
        ven = batting_score[(batting_score['batsman']==player) & (batting_score['venue']==j)]
        ven = ven.sort_values(by='match_id')
        ven['pre_venue1'] = ven['batsman_runs'].shift(1)
        ven['pre_venue2'] = ven['batsman_runs'].shift(2)
        ven['pre_venue3'] = ven['batsman_runs'].shift(3)
        venue_df = venue_df.append(ven)
    venue_df = venue_df[['match_id','batsman','Opp_Team','venue','pre_venue1','pre_venue2','pre_venue3']]
    bat2 = pd.merge(bat1,venue_df,left_on=['match_id','batsman','Opp_Team','venue'],right_on=['match_id','batsman','Opp_Team','venue'],how='left')
    bowl_opp_df = pd.DataFrame()
    for i in [i for i in teams if i not in own] :
        bowl_opp = batting_score[(batting_score['Own_Team']==own) & (batting_score['Opp_Team']==i)].groupby(by=['match_id','Opp_Team','venue'],as_index=False).sum()[['match_id','Opp_Team','venue','batsman_runs','out_bowler']]
        bowl_opp = bowl_opp.sort_values(by='match_id')
        bowl_opp['avg'] = bowl_opp['batsman_runs'] / bowl_opp['out_bowler'].apply(lambda x: 1 if x==0 else x)
        bowl_opp['bowl_opp1'] = bowl_opp['avg'].shift(1)
        bowl_opp['bowl_opp2'] = bowl_opp['avg'].shift(2)
        bowl_opp['bowl_opp3'] = bowl_opp['avg'].shift(3)
        bowl_opp_df = bowl_opp_df.append(bowl_opp)
    bat3 = pd.merge(bat2,bowl_opp_df,left_on=['match_id','Opp_Team','venue'],right_on=['match_id','Opp_Team','venue'],how='left')
    bat3 = bat3[['match_id', 'batsman', 'Own_Team', 'Opp_Team', 'venue','batsman_runs_x', 'out_bowler_x', 'pre1', 'pre2', 'pre3', 'pre_opp1',
           'pre_opp2', 'pre_opp3', 'pre_venue1', 'pre_venue2', 'pre_venue3','bowl_opp1', 'bowl_opp2','bowl_opp3']]
    bowl_df = pd.DataFrame()
    for i in [i for i in teams if i not in own] :
        bowl = batting_score[(batting_score['Opp_Team']==i)].groupby(by=['match_id','Opp_Team','venue'],as_index=False).sum()[['match_id','Opp_Team','venue','batsman_runs','out_bowler']]
        bowl = bowl.sort_values(by='match_id')
        bowl['avg'] = bowl['batsman_runs'] / bowl['out_bowler'].apply(lambda x: 1 if x==0 else x)
        bowl['bowl_pre1'] = bowl['avg'].shift(1)
        bowl['bowl_pre2'] = bowl['avg'].shift(2)
        bowl['bowl_pre3'] = bowl['avg'].shift(3)
        bowl_df = bowl_df.append(bowl)
        bowl_df.dropna(inplace=True)
    bat4 = pd.merge(bat3,bowl_df,left_on=['match_id','Opp_Team','venue'],right_on=['match_id','Opp_Team','venue'],how='left')
    bat4 = bat4[['match_id', 'batsman', 'Own_Team', 'Opp_Team', 'venue','batsman_runs_x', 'out_bowler_x', 'pre1', 'pre2', 'pre3', 'pre_opp1',
           'pre_opp2', 'pre_opp3', 'pre_venue1', 'pre_venue2', 'pre_venue3','bowl_pre1', 'bowl_pre2', 'bowl_pre3','bowl_opp1', 'bowl_opp2','bowl_opp3']]
    bowl_venue_df = pd.DataFrame()
    for i in [i for i in teams if i not in own] :
        for j in venues :
            bowl_venue = batting_score[(batting_score['Opp_Team']==i) & (batting_score['venue']==j)].groupby(by=['match_id','Opp_Team','venue'],as_index=False).sum()[['match_id','Opp_Team','venue','batsman_runs','out_bowler']]
            bowl_venue = bowl_venue.sort_values(by='match_id')
            bowl_venue['avg'] = bowl_venue['batsman_runs'] / bowl_venue['out_bowler'].apply(lambda x: 1 if x==0 else x)
            bowl_venue['venue_opp1'] = bowl_venue['avg'].shift(1)
            bowl_venue['venue_opp2'] = bowl_venue['avg'].shift(2)
            bowl_venue['venue_opp3'] = bowl_venue['avg'].shift(3)
            bowl_venue_df = bowl_venue_df.append(bowl_venue)
    bat5 = pd.merge(bat4,bowl_venue_df,left_on=['match_id','Opp_Team','venue'],right_on=['match_id','Opp_Team','venue'],how='left')
    bat5 = bat5[['Opp_Team', 'venue','batsman_runs_x', 'out_bowler_x', 
                 'pre1', 'pre2', 'pre3', 'pre_opp1','pre_opp2', 'pre_opp3', 'pre_venue1', 'pre_venue2', 
                 'pre_venue3','bowl_pre1', 'bowl_pre2', 'bowl_pre3','bowl_opp1', 'bowl_opp2','bowl_opp3',
                 'venue_opp1','venue_opp2', 'venue_opp3']]
    bat5['pre_opp1'].fillna(value=bat5[['pre1','pre2','pre3']].apply(fillna,axis=1),inplace=True)
    bat5['pre_opp2'].fillna(value=bat5['pre_opp1'],inplace=True)
    bat5['pre_opp3'].fillna(value=bat5[['pre_opp1','pre_opp2']].apply(fillna2,axis=1),inplace=True)

    bat5['pre_venue1'].fillna(value=bat5[['pre1','pre2','pre3']].apply(fillna,axis=1),inplace=True)
    bat5['pre_venue2'].fillna(value=bat5['pre_venue1'],inplace=True)
    bat5['pre_venue3'].fillna(value=bat5[['pre_venue1','pre_venue2']].apply(fillna2,axis=1),inplace=True)

    bat5['bowl_opp1'].fillna(value=bat5[['bowl_pre1','bowl_pre2','bowl_pre3']].apply(fillna,axis=1),inplace=True)
    bat5['bowl_opp2'].fillna(value=bat5['bowl_opp1'],inplace=True)
    bat5['bowl_opp3'].fillna(value=bat5[['bowl_opp1','bowl_opp2']].apply(fillna2,axis=1),inplace=True)

    bat5['venue_opp1'].fillna(value=bat5[['bowl_pre1','bowl_pre2','bowl_pre3']].apply(fillna,axis=1),inplace=True)
    bat5['venue_opp2'].fillna(value=bat5['venue_opp1'],inplace=True)
    bat5['venue_opp3'].fillna(value=bat5[['venue_opp1','venue_opp2']].apply(fillna2,axis=1),inplace=True)

    bat_dummies = pd.get_dummies(bat5,columns=['Opp_Team','venue'])
    bat_dummies.dropna(inplace=True)
    if bat_dummies.shape[0] >= 25 :
        bat_dummies['is_played']=bat_dummies[['batsman_runs_x','out_bowler_x']].apply(is_played,axis=1)
        X_train = bat_dummies.iloc[:-5,2:-1]
        y_train = bat_dummies.iloc[:-5,-1]
        X_test = bat_dummies.iloc[-5:,2:-1]
        print(X_test.columns)
        y_test = bat_dummies.iloc[-5:,-1]
        X_val = np.zeros(X_test.shape[1])
        X_val[0:3] = batting_score[(batting_score['batsman']=='KL Rahul')].sort_values(by='match_id').tail(3)['batsman_runs'].values
        X_val[3:6] = batting_score[(batting_score['batsman']=='KL Rahul') & 
                                   (batting_score['Opp_Team']=='KKR')].sort_values(by='match_id').tail(3)['batsman_runs'].values
        X_val[6:9] = batting_score[(batting_score['batsman']=='KL Rahul') & 
                                   (batting_score['venue']=='Punjab Cricket Association Stadium Mohali')].sort_values(by='match_id').tail(3)['batsman_runs'].values
        b1 = batting_score[(batting_score['Opp_Team']=='KKR')].groupby(by=['match_id','Opp_Team','venue'],as_index=False).sum()[['match_id','Opp_Team','venue','batsman_runs','out_bowler']].sort_values(by='match_id').tail(3)[['batsman_runs','out_bowler']]
        b1['avg'] = b1['batsman_runs'] / b1['out_bowler'].apply(lambda x: 1 if x==0 else x)
        X_val[9:12] = b1['avg'].values
        b2 = batting_score[(batting_score['venue']=='Punjab Cricket Association Stadium Mohali') & (batting_score['Opp_Team']=='KKR')].groupby(by=['match_id','Opp_Team','venue'],as_index=False).sum()[['match_id','Opp_Team','venue','batsman_runs','out_bowler']].sort_values(by='match_id').tail(3)[['batsman_runs','out_bowler']]
        b2['avg'] = b2['batsman_runs'] / b2['out_bowler'].apply(lambda x: 1 if x==0 else x)
        X_val[12:15] = b2['avg'].values
        b3 = batting_score[(batting_score['Own_Team']=='KXIP') & (batting_score['venue']=='Punjab Cricket Association Stadium Mohali')].groupby(by=['match_id','Opp_Team','venue'],as_index=False).sum()[['match_id','Opp_Team','venue','batsman_runs','out_bowler']].sort_values(by='match_id').tail(3)[['batsman_runs','out_bowler']]
        b3['avg'] = b3['batsman_runs'] / b3['out_bowler'].apply(lambda x: 1 if x==0 else x)
        X_val[15:18] = b3['avg'].values
        opposition = 'Opp_Team_'+ 'KKR'
        for i in range(0,len(X_train.columns)) :
            if opposition in X_train.columns[i] :
                opp_index = i
        X_val[opp_index] = 1
        match_venue = 'venue_'+ 'Punjab Cricket Association Stadium Mohali'
        for i in range(0,len(X_train.columns)) :
            if match_venue in X_train.columns[i] :
                venue_index = i
        X_val[venue_index] = 1        
        cls = GradientBoostingClassifier(n_estimators=int(X_train.shape[1]/2),random_state=0)
        cls.fit(X_train,y_train)
        pred = cls.predict_proba(X_test)[:,1]
        pred = pred > 0.5
        from sklearn.metrics import confusion_matrix,classification_report
        print(player + ' :')
        print(confusion_matrix(y_test,pred))
        print(cls.score(X_test,y_test))
```
