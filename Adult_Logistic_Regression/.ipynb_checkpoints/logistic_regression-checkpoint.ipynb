```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.linear_model import LogisticRegression
import warnings
from sklearn.model_selection import cross_val_score
from matplotlib import rcParams
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import confusion_matrix,classification_report,roc_curve,auc
```

```python
%matplotlib inline
warnings.filterwarnings('ignore')
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
rcParams['axes.titlepad'] = 20
```

```python
def clensing(df,cols) :
    df[cols] = df[cols].apply(lambda x: x.str.strip())
    df.drop_duplicates(inplace=True)
    df['Target'].replace(to_replace='<=50K',value='0',inplace=True)
    df['Target'].replace(to_replace='>50K',value='1',inplace=True)
    df['Target'] = df['Target'].astype(int)
    df.replace(to_replace='?',value=np.NaN,inplace=True)
    df.dropna(inplace=True)
    df.replace(regex=r'-',value='_',inplace=True)
    df.drop(labels='education_num',axis=1,inplace=True)
```

```python
def plotting_numerical(df,col,target,s):
    g = sns.FacetGrid(data=df,col=target,size=s,hue=target,margin_titles=True)
    g.map(sns.distplot,col)
    for ax in g.axes.flatten(): 
        for tick in ax.get_xticklabels(): 
            tick.set(rotation=90)
```

```python
def plotting_categorical(df,col,target,country,r,c,x,y):    
    fig, ax = plt.subplots(nrows=r,ncols=c,sharey=True,sharex=False)
    countries = df[country].unique()
    for i in range(0,len(countries)) :
        table=pd.crosstab(df[df['Country_US']==countries[i]][col],df[target])
        df2 = table.div(table.sum(1).astype(float), axis=0)
        df2.plot(kind='bar', stacked=True,ax=ax[i],figsize=(x,y),fontsize=11,colormap='Paired')
        ax[i].set_title('{} vs Income for {}'.format(col.capitalize(),countries[i]),fontsize=16,fontweight='bold',color='steelblue')
        ax[i].set_xlabel(col.capitalize(),fontsize=13,color='steelblue')
        ax[i].set_ylabel('Proportion of Customers',fontsize=13,color='steelblue')
        ax[i].legend(loc='center left', bbox_to_anchor=(1, 0.5))
    fig.subplots_adjust(hspace=1)
```

```python
def normalize(df_train,df_test,min_max_age,min_max_fnlwgt,min_max_capital_gain,min_max_capital_loss,min_max_hours_per_week):
    df_train['age_scale']=min_max_age.fit_transform(df_train[['age']])
    df_train['fnlwgt_scale']=min_max_fnlwgt.fit_transform(df_train[['fnlwgt']])
    df_train['capital_gain_scale']=min_max_capital_gain.fit_transform(df_train[['capital_gain']])
    df_train['capital_loss_scale']=min_max_capital_loss.fit_transform(df_train[['capital_loss']])
    df_train['hours_per_week_scale']=min_max_hours_per_week.fit_transform(df_train[['hours_per_week']])
    df_test['age_scale']=min_max_age.fit_transform(df_test[['age']])
    df_test['fnlwgt_scale']=min_max_fnlwgt.fit_transform(df_test[['fnlwgt']])
    df_test['capital_gain_scale']=min_max_capital_gain.fit_transform(df_test[['capital_gain']])
    df_test['capital_loss_scale']=min_max_capital_loss.fit_transform(df_test[['capital_loss']])
    df_test['hours_per_week_scale']=min_max_hours_per_week.fit_transform(df_test[['hours_per_week']])
```

```python
def fit_model(X_train,y_train,X_test,y_test) : 
    regression = LogisticRegression()
    model = regression.fit(X_train,y_train)
    accuracies = cross_val_score(model,X_train,y_train,cv=10)
    print('***** Accuracies ***** \n\n'+str(accuracies))
    print('\n\n***** Accuracies Mean ***** \n\n'+str(accuracies.mean()))
    predict = model.predict(X_test)
    cm = confusion_matrix(y_test,predict)
    cr = classification_report(y_test,predict)
    print('\n\n***** Confusion Matrix ***** \n\n'+str(cm))
    print('\n\n***** Classification Report ***** \n\n'+str(cr))
    probs = model.predict_proba(X_test)
    preds = probs[:,1]
    fpr, tpr, threshold = roc_curve(y_test, preds)
    roc_auc = auc(fpr, tpr)
    plt.title('Receiver Operating Characteristic')
    plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
    return model
```

```python
df_train = pd.read_csv('adult_train.csv',names=['age','workclass','fnlwgt','education','education_num',
                                          'marital_status','occupation','relationship','race','sex',
                                          'capital_gain','capital_loss','hours_per_week','native_country','Target'])
clensing(df_train,df_train.select_dtypes(include=['object']).columns)
```

```python
df_test = pd.read_csv('adult_test.csv',names=['age','workclass','fnlwgt','education','education_num',
                                          'marital_status','occupation','relationship','race','sex',
                                          'capital_gain','capital_loss','hours_per_week','native_country','Target'])
df_test['Target'].replace(to_replace=' <=50K.',value='<=50K',inplace=True)
df_test['Target'].replace(to_replace=' >50K.',value='>50K',inplace=True)
clensing(df_test,df_test.select_dtypes(include=['object']).columns)
```

```python
train_objs_num = len(df_train)
dataset = pd.concat(objs=[df_train, df_test], axis=0)
dataset = pd.get_dummies(dataset,drop_first=True)
train_data = dataset[:train_objs_num]
test_data = dataset[train_objs_num:]
```

```python
min_max_age =MinMaxScaler()
min_max_fnlwgt =MinMaxScaler()
min_max_capital_gain =MinMaxScaler()
min_max_capital_loss =MinMaxScaler()
min_max_hours_per_week =MinMaxScaler()
```

```python
normalize(train_data,test_data,min_max_age,
          min_max_fnlwgt,min_max_capital_gain,
          min_max_capital_loss,min_max_hours_per_week)
```

```python
X_train =train_data.iloc[:,6:]
y_train = train_data.iloc[:,5]
X_test =test_data.iloc[:,6:]
y_test = test_data.iloc[:,5]
model = fit_model(X_train,y_train,X_test,y_test)
```
